```python
df = pd.read_csv('new_cases_vlooked.csv',on_bad_lines='skip',sep=';')
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
df2 = df.groupby(['location','capital','lat','lon','year','month'])['new_cases'].sum()
df2
```

```python
df3 = pd.DataFrame(df2)
```

```python
df3 = df3.reset_index()
```

```python
df3['a']=df3['lat'].apply(lambda x: x[-1])
```

```python
def f_process(lat):
        lat = lat.replace(",", ".")
        if lat.endswith('N'):
            return float(lat.replace('N', ''))
        elif lat.endswith('B'):
            return float(lat.replace('B', ''))
        else:
            return -float(lat.replace('S', ''))

df3['lat'] = df3['lat'].apply(f_process)
```

```python
def f_process(lon):
        lon = lon.replace(",", ".")
        if lon.endswith('E'):
            return float(lon.replace('E', ''))
        else:
            return -float(lon.replace('W', ''))

df3['lon'] = df3['lon'].apply(f_process)
```
